```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/data.csv')
```

###DATA CLEANING

```python
#Fillna
#dropna
df.dropna(how='any', inplace=True)
```

```python
len(df)
```

```python
#Checking y variable
df.diagnosis.value_counts()
```

```python
#checking precentage
df.diagnosis.value_counts()/len(df)*100
```

#Splitting the Data

```python
X = df.drop(["id","diagnosis"], axis=1)
```

```python
y = df.diagnosis.values
```

#TRAINING AND TEST SPLIT

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.25)
```

```python
len(X_train)
```

```python
len(X_test)
```

#FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_sc
```

#MODEL BUILDING

```python
from sklearn.neighbors import KNeighborsClassifier

#intiate the classifier
model_knn = KNeighborsClassifier()

#Passing the data to Classifier
model_knn.fit(X_train_sc, Y_train)
```

```python
y_pred_knn = model_knn.predict(X_test_sc)
```

```python
y_pred_knn
```

```python
Y_test
```

```python
 #Classification metrics to check how to model behaves

 from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
 print(np.round(confusion_matrix(Y_test, y_pred_knn)*100,2))
 print(classification_report(Y_test, y_pred_knn))
 print(np.round(accuracy_score(Y_test, y_pred_knn)*100,2))
```
